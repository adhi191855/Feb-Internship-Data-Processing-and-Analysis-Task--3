**1.Employee Performance Bonus Eligibility**

```python
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}
#Find the highest score
highest_score=0
for score in employees.values():
  if score>highest_score:
    highest_score=score

#Find employee with highest score
top_performers=[]
for name in employees:
  if employees[name]==highest_score:
    top_performers.append(name)
print("Top Performers Eligible for bonus:",",".join(top_performers),"(Score:",highest_score,")")
```

**2.Search Query Keyword Analysis**

```python
query = "Buy mobile phone buy phone online"
#Convert query to lower case
query=query.lower()
#split query
words=query.split()
#Count word frequency
word_count={}
for word in words:
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word]=1

#Dictionary to store repeated words
result={}
for word in word_count:
  if word_count[word]>1:
    result[word]=word_count[word]
print(result)
```

**3.Sensor Data Validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings=[]

for index,value in enumerate(sensor_readings):
  if value%2==0:
    valid_readings.append((index,value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**4.Email Domain Usage Analysis**

```python
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_count={}
#Count email domains
for email in emails:
  domain=email.split("@")[1]
  if domain in domain_count:
    domain_count[domain]+=1
  else:
    domain_count[domain]=1
total_emails=len(email)
for domain in domain_count:
  percentage=(domain_count[domain]/total_emails)*100
  print(domain,":",int(percentage),"%")
```

**5.Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Detect spikes (more than 30% above average)
threshold = average_sales * 1.3

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

### 6.Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

# Count occurrences
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicates only
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
